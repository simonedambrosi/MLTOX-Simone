```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


species = pd.read_csv('C:/Users/Simone/Desktop/data/species.txt', sep = '\|', engine = 'python')
print(species.shape)
# species.head()
```

```python
# Se non hanno l'ecotox_group
species = species[~species.ecotox_group.isnull()]
print(species.shape)
# species.head()
```

```python
species_fish = species[species.ecotox_group.str.contains("Fish")]
print(species_fish.shape)
# species_fish.head()
```

```python
def null_output_counts(dataframe):
    
    # Find columns that start with the interesting feature
    features_interested = list(dataframe.columns)
    
    df_nan  = pd.DataFrame (index = features_interested, columns = ['null_values_inc_NC_NR%', '#outputs'])
    
    #Count total NaN + NR + NC
    for i in features_interested:
        df_nan['null_values_inc_NC_NR%'][i] = (sum(dataframe[i].isnull()) + len(dataframe[dataframe[i] == "NR"]) + len(dataframe[dataframe[i] =="NC"]))/len(dataframe)*100
        df_nan['#outputs'][i] = len(dataframe[i].unique())
    return df_nan
```

```python
noc_species = null_output_counts(species_fish)
noc_species
```

```python
# filtro sulle caratteristiche dei pesci che hanno meno della met√† di valori mancanti
noc_species = noc_species[noc_species['null_values_inc_NC_NR%'] < 50]
noc_species = noc_species[(noc_species['#outputs'] > 1) & (noc_species['#outputs'] < species_fish.shape[0]/2)]
noc_species
```

```python
# Ecotox_group Not useful
noc_species.drop(index = ['ecotox_group'], inplace =True)
species_fish = species_fish[noc_species.index.tolist()]
print(species_fish.shape)
species_fish
```

```python
def checkTree(dataset):
    '''
    Vogliamo vedere se nel dataset delle specie di animali su cui sono stati effettuati i test √® presente un albero 
    filogenetico, ossia se le specie di pesci condividono il genere allora condividono anche la famiglia, 
    se condividono la famiglia allora condividono anche l'ordine tassonomico cos√¨ via fino alla classe.
    '''
    lst_fam =list()
    lst_to = list() 
    lst_c = list()  # le verifiche incrociate
    
    for r1 in dataset.values:
        for r2 in dataset.values:
            if r1[3] == r2[3]:
                lst_fam.append(r1[2] == r2[2])
            if r1[2] == r2[2]:
                lst_to.append(r1[1] == r2[1])
            if r1[1] == r2[1]:
                lst_c.append(r1[0] == r2[0])
    
    lst_fam = pd.Series(lst_fam)
    lst_to = pd.Series(lst_to)
    lst_c = pd.Series(lst_c)
    
    if (sum(lst_fam == False) == 0) & (sum(lst_to == False) == 0) & (sum(lst_c == False) == 0):
        print('Tutto ok, √® un albero')
    else:
        print('Ops... non √® un albero')

checkTree(species_fish.drop(columns = ['genus']))
```

```python
class_ = species_fish['class'].unique()
tax_order = species_fish['tax_order'].unique()
family = species_fish['family'].unique()
genus_sp = species_fish['gen_spec'].unique()


c_to = dict()
to_fam = dict()
fam_gensp = dict()


for c in class_:
    c_to[c] = list()

for to in tax_order:
    to_fam[to] = list()
    
for f in family:
    fam_gensp[f] = list()

    
for i in species_fish.values:
    if i[1] not in c_to[i[0]]:
        c_to[i[0]].append(i[1])
    
    if i[2] not in to_fam[i[1]]:
        to_fam[i[1]].append(i[2])
    
    if i[3] not in fam_gensp[i[2]]:
        fam_gensp[i[2]].append(i[3])


albero = c_to.copy()
albero.update(to_fam)
albero.update(fam_gensp)    
```

```python
import networkx as nx
import matplotlib.pyplot as plt

alnx = nx.Graph(albero)

plt.figure(figsize=(30,30))
nx.draw(alnx, pos= nx.spring_layout(alnx), node_list= class_) # figura per la tesi NON CAMBIARE
```

```python
from ete3 import Tree, TreeStyle


t = Tree(new_tree, format = 8)
ts = TreeStyle()
ts.show_leaf_name = True
ts.mode = "c"
ts.arc_start = -360 # 0 degrees = 3 o'clock
ts.arc_span = 360
t.show(tree_style=ts)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def import_raw_data(DATA_PATH_TESTS, DATA_PATH_RESULTS, DATA_PATH_SPECIES):
    tests = pd.read_csv(DATA_PATH_TESTS, sep = '\|', engine = 'python')
    print('tests loaded')
    species = pd.read_csv(DATA_PATH_SPECIES, sep = '\|', engine = 'python')
    print('species loaded')
    results = pd.read_csv(DATA_PATH_RESULTS, sep = '\|', engine = 'python')
    print('results loaded')

    return tests, species, results

path_tests = 'C:/Users/Simone/Desktop/data/tests.txt'
path_species = 'C:/Users/Simone/Desktop/data/species.txt'
path_results = 'C:/Users/Simone/Desktop/data/results.txt'

tests, species, results = import_raw_data(path_tests,path_results, path_species)
```

```python
# Filtro sugli endpoint 
resc = results[(results.endpoint.str.contains('EC50')) | (results.endpoint.str.contains('LC50'))]
# Filtro sull'effetto
resc_mor = resc[resc.effect.str.contains('MOR')]

# Filtro sulle specie
species = species[~species.ecotox_group.isnull()]
species_fish = species[species.ecotox_group.str.contains("Fish")]

# No Embrions
test_no_EM = tests[tests.organism_lifestage != "EM"]

# Unisco le informazioni sui pesci con gli esperimenti
test_fish_only = test_no_EM.merge(species_fish, on="species_number")

# Unisco le informazioni sui risultati degli esperimenti con gli esperimenti
results_prefilter = resc_mor.merge(test_fish_only, on = "test_id")
```

```python
def null_output_counts(dataframe):
    
    # Find columns that start with the interesting feature
    features_interested = list(dataframe.columns)
    
    df_nan  = pd.DataFrame (index = features_interested, columns = ['null_values_inc_NC_NR%', '#outputs'])
    
    #Count total NaN + NR + NC
    for i in features_interested:
        df_nan.loc[i,'null_values_inc_NC_NR%'] = (sum(dataframe[i].isnull()) + len(dataframe[dataframe[i] == "NR"]) + len(dataframe[dataframe[i] =="NC"]))/len(dataframe)*100
        df_nan.loc[i,'#outputs'] = len(dataframe[i].unique())
    return df_nan

df_noc = null_output_counts(results_prefilter)
```
